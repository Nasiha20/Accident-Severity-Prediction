## Data preprocessing

#### Label encoding

```python
from sklearn.preprocessing import LabelEncoder
severity_le=LabelEncoder()
accidents['Severity']=severity_le.fit_transform(accidents['Severity'])
accidents.head()
```

#### Min Max Normalization

```python
accidents['Safety_Score'] = (accidents['Safety_Score']-accidents['Safety_Score'].min())/(accidents['Safety_Score'].max()-accidents['Safety_Score'].min())
accidents['Days_Since_Inspection'] = (accidents['Days_Since_Inspection']-accidents['Days_Since_Inspection'].min())/(accidents['Days_Since_Inspection'].max()-accidents['Days_Since_Inspection'].min())
accidents['Total_Safety_Complaints'] = (accidents['Total_Safety_Complaints']-accidents['Total_Safety_Complaints'].min())/(accidents['Total_Safety_Complaints'].max()-accidents['Total_Safety_Complaints'].min())
accidents['Control_Metric'] = (accidents['Control_Metric']-accidents['Control_Metric'].min())/(accidents['Control_Metric'].max()-accidents['Control_Metric'].min())
accidents['Turbulence_In_gforces'] = (accidents['Turbulence_In_gforces']-accidents['Turbulence_In_gforces'].min())/(accidents['Turbulence_In_gforces'].max()-accidents['Turbulence_In_gforces'].min())
accidents['Cabin_Temperature'] = (accidents['Cabin_Temperature']-accidents['Cabin_Temperature'].min())/(accidents['Cabin_Temperature'].max()-accidents['Cabin_Temperature'].min())
accidents['Accident_Type_Code'] = (accidents['Accident_Type_Code']-accidents['Accident_Type_Code'].min())/(accidents['Accident_Type_Code'].max()-accidents['Accident_Type_Code'].min())
accidents['Max_Elevation'] = (accidents['Max_Elevation']-accidents['Max_Elevation'].min())/(accidents['Max_Elevation'].max()-accidents['Max_Elevation'].min())
accidents['Violations'] = (accidents['Violations']-accidents['Violations'].min())/(accidents['Violations'].max()-accidents['Violations'].min())
accidents['Adverse_Weather_Metric'] = (accidents['Adverse_Weather_Metric']-accidents['Adverse_Weather_Metric'].min())/(accidents['Adverse_Weather_Metric'].max()-accidents['Adverse_Weather_Metric'].min())
accidents['Accident_ID'] = (accidents['Accident_ID']-accidents['Accident_ID'].min())/(accidents['Accident_ID'].max()-accidents['Accident_ID'].min())

accidents.head()
```

### Algorithms to find optimal number of clusters

### Elbow Method

```python
from sklearn.cluster import KMeans
kmeans_foreach_k = [KMeans(n_clusters=k, random_state=42).fit(x2)
                for k in range(1, 10)]
inertias_foreach_k = [model.inertia_ for model in kmeans_foreach_k]

plt.figure(figsize=(8, 3.5))
plt.plot(range(1, 10), inertias_foreach_k, "b-*")
plt.xlabel("$k$", fontsize=14)
plt.ylabel("Inertia", fontsize=14)
plt.show()
```

### Silhouette Analysis

```python
from yellowbrick.cluster import SilhouetteVisualizer
fig, ax = plt.subplots(3, 2, figsize=(15,15))
for i in [2, 3, 4, 5, 6, 7]:
    silhouette_km = KMeans(n_clusters=i, init='k-means++', n_init=10, max_iter=100, random_state=5)
    q, mod = divmod(i, 2)
    accidents_visualizer = SilhouetteVisualizer(silhouette_km, colors='yellowbrick', ax=ax[q-1][mod])
    accidents_visualizer.fit(x2)
    
ax[0,0].title.set_text("silhouette analysis for k = 2")
ax[0,1].title.set_text("silhouette analysis for k = 3")
ax[1,0].title.set_text("silhouette analysis for k = 4")
ax[1,1].title.set_text("silhouette analysis for k = 5")
ax[2,0].title.set_text("silhouette analysis for k = 6")
ax[2,1].title.set_text("silhouette analysis for k = 7")
```

### Knee Point Detection

```python
from kneed import KneeLocator
sse_values = {}
increment_counter = 0
decrement_counter = 0
for k in range(1, 10):
        kmeans_knee = KMeans(n_clusters=k, random_state=1)
        kmeans_knee.fit(x2)
        sse_values[k] = kmeans_knee.inertia_   
knee_point = KneeLocator(x=list(sse_values.keys()), y=list(sse_values.values()), curve='convex', direction='decreasing')
k = knee_point.knee + increment_counter - decrement_counter
k
```

#### Based on the above methods, the optimal number of clusters is 4
